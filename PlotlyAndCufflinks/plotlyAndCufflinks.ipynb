```python
import plotly.express as px
import seaborn as sns

tips = sns.load_dataset('tips')

fig = px.line(
    tips,
    y='total_bill',
    title='Total Bill'
)

fig.show()
```

```python
tips.groupby('day')['tip'].mean().iplot(kind = 'bar')
```

```python
import plotly.express as px

avg_tip = tips.groupby('day')['tip'].mean().reset_index()

fig = px.bar(
    avg_tip,
    x='day',
    y='tip',
    title='Average Tip by Day'
)

fig.show()
```

```python
tips.iplot(kind='scatter',x ='total_bill',y = 'tip',mode = 'markers')
```

```python
import plotly.express as px

fig = px.scatter(
    tips,
    x='total_bill',
    y='tip',
    title='Total Bill vs Tip'
)

fig.show()
```

```python
tips[['total_bill','day']].iplot(kind = 'box',x = 'total_bill',y = 'day')
```

```python
sns.boxplot(
    x='day',
    y='total_bill',
    data=tips
)
```

```python
tips[['total_bill','tip']].iplot(kind='box')
```

```python
import plotly.express as px
import seaborn as sns

tips = sns.load_dataset('tips')

fig = px.box(
    tips,
    x='day',
    y='total_bill',
    title='Total Bill Distribution by Day'
)

fig.show()
```
