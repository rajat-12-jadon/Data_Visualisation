Object Oriented Way


Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.
    
figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).m

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1, 0.1, 1, 1])
axis1.plot(x, y)
axis2 = fig.add_axes([0.2, 0.6, 0.5, 0.5])
axis2.plot(y, x)
```

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1, 0.1, 1, 1])
axis1.plot(x, y)
axis2 = fig.add_axes([0.5, 0.2, 0.5, 0.5])
axis2.plot(y, x)
```

```python
fig = plt.figure(figsize=(20,8),dpi = 500)
axis1 = fig.add_axes([0.1,0.1,.4,.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,.3,.3])
axis2.plot(x,y)
```
